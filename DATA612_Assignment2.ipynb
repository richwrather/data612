```python
# Read in the stocks_yahoo.csv dataset from the Franklin GitHub repo.
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/frankData612/data_612/master/stock_data/stocks_yahoo.csv")
```

```python
# Print the first 5 records of the stocks_yahoo dataset.
df.head()
```

```python
# Print the information about the dataset including the data types
#   Notice that the 'date' field is of type 'object'.
df.info()
```

###1. Find the maximum date found in the dataset.

```python
# Convert the 'date' field into a datetime64 type
entry_date = pd.to_datetime(df['date'], format='%b %d %Y')

print(entry_date)

```

```python
# Find the most recent date by calling the max() function
max_entry_date = entry_date.max()

print(max_entry_date)
```

###2. Subtract all the dates on the column from the maximum date

```python
# Calculate the difference between the max month and the month in each record.
days_diff = max_entry_date - entry_date
days_diff 
```

###3. Convert the number of days into number of months.

```python
# Convert the difference in days into the number of months by using the 
#   Numpy timedelta64() function.

import numpy as np

months = (max_entry_date - entry_date) / np.timedelta64(1, 'M')
months
```

###4. Save the result on a new column and add to the data set.

```python
# Create a new column named 'months_diff' and write the number of months between 
#   the max date and the date for each record.
 
df['months_diff'] = months
df
```

###5. Save the dataset as a csv file.

```python
# Write the new dataframe as a csv file.

df.to_csv('data612_stocks_yahoo.csv', index=False)
```
