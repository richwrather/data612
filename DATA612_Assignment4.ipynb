```python
import pandas as pd

# Read in the stocks_yahoo.csv dataset from the Franklin GitHub repo.
df = pd.read_csv("https://raw.githubusercontent.com/frankData612/data_612/master/stock_data/stocks_yahoo.csv")

# Manually create a new data frame to be added to the original 
new_row_df = pd.DataFrame([[df.shape[0] + 1, 'Jun 13 2021', 'AAPL - Apple Inc.', 
                            '129.64', '-0.30 (-0.23%)', '130.01', '129.94', '129.87', '129.64 - 131.32', 
                            '119.02 - 141.56', '2.876813+07', '25813040', '1.178T', '1.25', '22.08', '11.93', 'NaN', '3.08 (1.16%)', '259.12']],
                columns=['no', 'date', 'company_name', 'price_at_close', 'price_change',
                         'price_after_hours', 'previous_close', 'today opened', 'price range',
                         '52 week range', 'volume', 'average volume', 'market cap',
                         'beta (3Y Monthly)', 'PE Ratio', 'EPS', 'earnings date',
                         'Dividend Yield', '1Y target est'])

# Concatenate the two data frames and ignore the index
new_df = pd.concat([df, new_row_df], ignore_index=True)
new_df
```

```python
# count the number of non-missing values
print(new_df.count())
```

```python
# Count the number of rows
num_rows = new_df.shape[0]

# Compute the number of missing rows
num_missing = num_rows - new_df.count()
print(num_missing)
```

```python
# Drop the rows with missing values
new_df_dropna = new_df.dropna()
new_df_dropna
```
