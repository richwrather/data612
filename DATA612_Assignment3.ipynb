###1. Create three different meaningful charts using Seaborn package on your selected data set.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read in the stocks_yahoo.csv dataset from the Franklin GitHub repo.
df = pd.read_csv("https://raw.githubusercontent.com/frankData612/data_612/master/stock_data/stocks_yahoo.csv")

# Convert the 'date' field into a datetime64 type
entry_date = pd.to_datetime(df['date'], format='%b %d %Y')
df["entry_date"] = entry_date
```

```python
# Subset the Apple data into a new Data Frame and order it in decending order by date
apple = df[df["company_name"] == "AAPL - Apple Inc."].sort_values(by="date", ascending=False)

# Use a Line Plot to display the change in Apple stock price
ax = plt.subplot()
ax = sns.lineplot(data=apple, x="entry_date", y="price_at_close")
ax.set_title("Apple Closing Price")
ax.set_xlabel("Date")
ax.set_ylabel("Price")
plt.show()
```

```python
# Create a new Data Frame with just the data from April 28, 2020
april282020 = df[df["entry_date"] == "2020-04-28"].sort_values(by="price_at_close", ascending=False).dropna()

# Setup a Barplot to show the top 5 companies based on stock price that day
ax = plt.subplot()
ax = sns.barplot(data=april282020.head(), x="company_name", y="price_at_close")
ax.set_title("Top 5 Closing Prices - April 28,2020")
ax.set_xlabel("Company Name")
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
ax.set_ylabel("Price")
plt.show()
```

```python
# Add a new column that contains the daily change in value
df["daily_return"] = df["previous_close"] - df["price_at_close"]

# Subset the Apple data into a new Data Frame
apple = df[df["company_name"] == "AAPL - Apple Inc."].sort_values(by="date", ascending=False)

# Use a Histogram to plot the daily returns into bins
sns.displot(apple['daily_return'].dropna(),bins=50,color='red')
```
